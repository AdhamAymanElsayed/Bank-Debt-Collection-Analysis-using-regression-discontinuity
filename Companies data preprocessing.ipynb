```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
# rows that have almost values are unique then, not affected the data
df.drop(columns=['Unnamed: 0.1','id','normalized_name','permalink','twitter_username',
                'homepage_url','entity_id','lng','lat'], inplace= True)
```

```python
df.head().T
```

```python
df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce')
current_year = dt.datetime.now().year
df['company_age'] = current_year - df['founded_at'].dt.year
```

```python
# Percantage of null Values 
((df.isna().sum()) / df.shape[0])*100
```

```python
# Columns wih null Values Greater than 70% will be dropped
df.drop(columns= ['ROI','funding_total_usd','funding_rounds','last_funding_at',
                  'first_investment_at','last_investment_at','investment_rounds','invested_companies'
                  ,'first_funding_at','state_code','parent_id','closed_at','short_description','founded_at'],inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#drop it , not affected the model becauese only one type in data
df.drop(columns=['entity_type'],inplace=True)
```
